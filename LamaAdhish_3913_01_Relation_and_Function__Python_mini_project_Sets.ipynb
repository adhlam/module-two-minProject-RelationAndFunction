**Dictionaries**

Lets define the dictionary that creates the relationship between the set A and B.

A = {x:x is a set of composite number; 0<x<16}

B = {x:x is a set of odd number; 0<x<16}

```python
A = ['4','6','8','9','10','12','14','15']
B = ['1','3','5','7','9','11','13','15']
dictionary = dict({a:b for a,b in zip(A, B)})
```

```python
actualDictionary = dict({'4':'5','9':'11', '12':'13'})
```

**Finding a new range**

```python
newlist2 = [actualDictionary.get(x,x) for x in A]
```

**Lets us now analyze the newlist**

The dictionary is a function. Though it is one-to-one but not onto and hence the function is not bijective.

The relation is not endo relation.
